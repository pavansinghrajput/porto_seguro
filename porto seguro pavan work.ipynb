->loading the recquired packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

->importing data

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#checking for null vlues in terms of -1
(df==-1).sum()
```

```python
#Analysis of target variable
df['Bin_target'] = df['target'].map({0: '0', 1: '1'})
sns.countplot(df.Bin_target);
plt.xlabel('x');
plt.ylabel('y');
plt.show()
```

```python
#separating the columns into  numeric , categorical , binary
features = df.columns.tolist()
features.remove('target')
# Numeric Features
numeric_features = [x for x in features if x[-3:] not in ['bin', 'cat']]
# Categorical Features
categorical_features = [x for x in features if x[-3:]=='cat']
# Binary Features
binary_features = [x for x in features if x[-3:]=='bin']
```

```python
#cor-rleation of numeric columns
cor_matrix = df[numeric_features].corr().round(2)
fig = plt.figure(figsize=(18,18));
sns.heatmap(cor_matrix, annot=True, center=0, cmap = sns.diverging_palette(250, 10, as_cmap=True), ax=plt.subplot(111));
plt.show()
```

```python
##cor-rleation of categorical columns
cor_matrix = df[categorical_features ].corr().round(2)
fig = plt.figure(figsize=(18,18));
sns.heatmap(cor_matrix, annot=True, center=0, cmap = sns.diverging_palette(250, 10, as_cmap=True), ax=plt.subplot(111));
plt.show()
```

```python
##cor-rleation of binary columns
cor_matrix = df[binary_features].corr().round(2)
fig = plt.figure(figsize=(18,18));
sns.heatmap(cor_matrix, annot=True, center=0, cmap = sns.diverging_palette(250, 10, as_cmap=True), ax=plt.subplot(111));
plt.show()
```
